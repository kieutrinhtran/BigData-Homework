### Demo tính năng MaxVersion

```python
import happybase
import time

# Kết nối đến HBase qua Thrift server
conn = happybase.Connection('localhost', port=9090)
conn.open()

# Xóa bảng nếu đã tồn tại
if b'Employee' in conn.tables():
    conn.disable_table('Employee')
    conn.delete_table('Employee')

# Tạo bảng mới với max_versions = 3
conn.create_table(
    'Employee',
    {'EmpID': dict(max_versions=3)}  # Chỉ giữ tối đa 3 phiên bản
)

table = conn.table('Employee')

# Ghi nhiều phiên bản vào cùng 1 ô (EmpID:code)
for i in range(1, 6):  # Ghi 5 version
    table.put(b'row1', {b'EmpID:code': f'E00{i}'.encode()})
    print(f"Ghi version {i}: E00{i}")
    time.sleep(1)  # Đảm bảo timestamp khác nhau

print("\n📜 Các phiên bản còn lại trong EmpID:code (sau khi giới hạn):")
# Đọc lại tất cả các phiên bản (bao gồm timestamp)
versions = table.cells(b'row1', b'EmpID:code', include_timestamp=True)

# In ra từng phiên bản
for idx, (value, ts) in enumerate(versions):
    print(f"Version {idx+1}: {value.decode()} @ timestamp {ts}")

conn.close()
```

```python
#test thêm in ra timestamp

import happybase
import time

# Kết nối đến HBase qua Thrift server
conn = happybase.Connection('localhost', port=9090)
conn.open()

# Xóa bảng nếu đã tồn tại
if b'Employee' in conn.tables():
    conn.disable_table('Employee')
    conn.delete_table('Employee')

# Tạo bảng mới với max_versions = 3
conn.create_table(
    'Employee',
    {'EmpID': dict(max_versions=3)}  # Chỉ giữ tối đa 3 phiên bản
)

table = conn.table('Employee')

# Ghi nhiều phiên bản vào cùng 1 ô (EmpID:code)
for i in range(1, 6):  # Ghi 5 version
    timestamp = int(time.time() * 1000)  # Lấy timestamp hiện tại (theo mili giây)
    table.put(b'row1', {b'EmpID:code': f'E00{i}'.encode()}, timestamp=timestamp)
    print(f"Ghi version {i}: E00{i} với timestamp {timestamp}")
    time.sleep(1)  # Đảm bảo timestamp khác nhau

print("\n📜 Các phiên bản còn lại trong EmpID:code (sau khi giới hạn):")
# Đọc lại tất cả các phiên bản (bao gồm timestamp)
versions = table.cells(b'row1', b'EmpID:code', include_timestamp=True)

# In ra từng phiên bản
for idx, (value, ts) in enumerate(versions):
    print(f"Version {idx+1}: {value.decode()} @ timestamp {ts}")

conn.close()
```

```python
#test đổi timestamp

import happybase
import time
from datetime import datetime

# Kết nối đến HBase qua Thrift server
conn = happybase.Connection('localhost', port=9090)
conn.open()

# Xóa bảng nếu đã tồn tại
if b'Employee' in conn.tables():
    conn.disable_table('Employee')
    conn.delete_table('Employee')

# Tạo bảng mới với max_versions = 3
conn.create_table(
    'Employee',
    {'EmpID': dict(max_versions=3)}  # Chỉ giữ tối đa 3 phiên bản
)

table = conn.table('Employee')

# Ghi nhiều phiên bản vào cùng 1 ô (EmpID:code)
for i in range(1, 6):  # Ghi 5 version
    timestamp = int(time.time() * 1000)  # Lấy timestamp hiện tại (theo mili giây)

    # Chuyển timestamp thành ngày giờ
    dt_object = datetime.utcfromtimestamp(timestamp / 1000)  # Chia cho 1000 vì timestamp ở mili giây
    formatted_time = dt_object.strftime('%Y-%m-%d %H:%M:%S')  # Định dạng theo kiểu YYYY-MM-DD HH:MM:SS

    table.put(b'row1', {b'EmpID:code': f'E00{i}'.encode()}, timestamp=timestamp)
    print(f"Ghi version {i}: E00{i} với timestamp {formatted_time}")

    time.sleep(1)  # Đảm bảo timestamp khác nhau

print("\n📜 Các phiên bản còn lại trong EmpID:code (sau khi giới hạn):")
# Đọc lại tất cả các phiên bản (bao gồm timestamp)
versions = table.cells(b'row1', b'EmpID:code', include_timestamp=True)

# In ra từng phiên bản với timestamp là ngày giờ
for idx, (value, ts) in enumerate(versions):
    dt_object = datetime.utcfromtimestamp(ts / 1000)  # Chia cho 1000 vì timestamp ở mili giây
    formatted_time = dt_object.strftime('%Y-%m-%d %H:%M:%S')
    print(f"Version {idx+1}: {value.decode()} @ timestamp {formatted_time}")

conn.close()
```

### Hãy tạo bảng Movie_MSSV (MSSV: mã số SV) với 2 family cột là:

```python
import happybase

# Tên bảng với MSSV của bạn
table_name = 'Movie_31221025966'

# Kết nối đến HBase qua Thrift server
conn = happybase.Connection('localhost', port=9090)
conn.open()

# Xóa bảng nếu đã tồn tại
if table_name.encode() in conn.tables():
    conn.disable_table(table_name)
    conn.delete_table(table_name)

# Tạo bảng mới với 2 column family: CoBan và ChiTiet
conn.create_table(
    table_name,
    {
        'CoBan': dict(),    # Thông tin cơ bản như tiêu đề, năm, thể loại
        'ChiTiet': dict()   # Thông tin chi tiết như đạo diễn, mô tả, diễn viên
    }
)

print(f"✅ Đã tạo bảng '{table_name}' thành công với 2 column family: 'CoBan' và 'ChiTiet'.")

conn.close()
```

```python
#Bước 3: Đọc file & import vào HBase
import happybase
import csv

# Kết nối đến HBase qua Thrift
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# Đọc file CSV đã làm sạch
with open('netflix_clean.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    count = 0

    for row in reader:
        if len(row) < 12:
            continue  # bỏ qua dòng thiếu cột

        row_key = f"r{count}".encode()

        # Ghi dữ liệu vào HBase, chia cột theo 2 column family
        table.put(row_key, {
            b'CoBan:show_id': row[0].encode(),
            b'CoBan:type': row[1].encode(),
            b'CoBan:title': row[2].encode(),

            b'ChiTiet:director': row[3].encode(),
            b'ChiTiet:cast': row[4].encode(),
            b'ChiTiet:country': row[5].encode(),
            b'ChiTiet:date_added': row[6].encode(),

            b'CoBan:release_year': row[7].encode(),
            b'CoBan:rating': row[8].encode(),
            b'CoBan:duration': row[9].encode(),

            b'ChiTiet:listed_in': row[10].encode(),
            b'ChiTiet:description': row[11].encode(),
        })

        count += 1
        if count >= 1000:  # Giới hạn 1000 dòng
            break

print(f"✅ Đã import {count} dòng dữ liệu vào bảng Movie_31221025966 với đầy đủ cột.")
conn.close()
```

#### Hãy hiển thị 100 dữ liệu từ s100 với chỉ 2 cột của CoBan và 2 cột của ChiTiet

```python
import happybase

# Kết nối đến HBase
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# Bước 1: Lấy toàn bộ dữ liệu có row_key là r***
rows = []
for key, data in table.scan():
    key_str = key.decode()
    if key_str.startswith('r') and key_str[1:].isdigit():
        rows.append((key_str, data))

# Bước 2: Sắp xếp theo số
rows.sort(key=lambda x: int(x[0][1:]))

# Bước 3: Lọc từ r100 trở đi
rows_from_r100 = [row for row in rows if int(row[0][1:]) >= 100]

# Bước 4: In tiêu đề
print(f"{'Row Key':<8} | {'Title':<30} | {'Year':<6} | {'Director':<20} | {'Country':<20}")
print("-" * 100)

# Bước 5: Hiển thị 100 dòng từ r100 với 2 cột CoBan + 2 cột ChiTiet
for idx, (key, data) in enumerate(rows_from_r100[:100]):
    title = data.get(b'CoBan:title', b'N/A').decode('utf-8', errors='ignore')
    year = data.get(b'CoBan:release_year', b'N/A').decode('utf-8', errors='ignore')
    director = data.get(b'ChiTiet:director', b'N/A').decode('utf-8', errors='ignore')
    country = data.get(b'ChiTiet:country', b'N/A').decode('utf-8', errors='ignore')

    print(f"{key:<8} | {title[:30]:<30} | {year:<6} | {director[:20]:<20} | {country[:20]:<20}")

conn.close()
```

```python
import happybase

# Kết nối
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# In tiêu đề
print(f"{'Row Key':<8} | {'Title':<30} | {'Year':<6}")
print("-" * 50)

# Duyệt và lọc bằng so sánh chuỗi
for key, data in table.scan():
    try:
        title = data.get(b'CoBan:title', b'N/A').decode('utf-8', errors='ignore')
        year = data.get(b'CoBan:release_year', b'').decode('utf-8', errors='ignore')

        if year > "2019":  # so sánh chuỗi
            print(f"{key.decode():<8} | {title[:30]:<30} | {year:<6}")
    except:
        continue

conn.close()
```

```python
import happybase

# Kết nối
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# Tiêu đề
print(f"{'Row Key':<8} | {'Title':<30} | {'Type':<10} | {'Country':<20}")
print("-" * 80)

# Duyệt bảng
for key, data in table.scan():
    try:
        title = data.get(b'CoBan:title', b'N/A').decode('utf-8', errors='ignore')
        type_ = data.get(b'CoBan:type', b'').decode('utf-8', errors='ignore')
        country = data.get(b'ChiTiet:country', b'').decode('utf-8', errors='ignore')

        # Điều kiện lọc
        if "United States" in country and type_ == "Movie":
            print(f"{key.decode():<8} | {title[:30]:<30} | {type_[:10]:<10} | {country[:20]:<20}")
    except:
        continue

conn.close()
```

##### Film trong cột listed_in với loại là Dramas

```python
import happybase

# Kết nối đến HBase
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# In tiêu đề
print(f"{'Row Key':<8} | {'Title':<30} | {'Genres':<30}")
print("-" * 70)

# Duyệt bảng và lọc theo điều kiện: listed_in chứa "Dramas"
for key, data in table.scan():
    try:
        title = data.get(b'CoBan:title', b'N/A').decode('utf-8', errors='ignore')
        genres = data.get(b'ChiTiet:listed_in', b'').decode('utf-8', errors='ignore')

        if "Dramas" in genres:
            print(f"{key.decode():<8} | {title[:30]:<30} | {genres[:30]:<30}")
    except:
        continue

conn.close()
```

#### Kết hợp cả 3 điều kiện

```python
import happybase

# Kết nối đến HBase
conn = happybase.Connection('localhost', port=9090)
conn.open()
table = conn.table('Movie_31221025966')

# In tiêu đề
print(f"{'Row Key':<8} | {'Title':<30} | {'Year':<6} | {'Type':<10} | {'Country':<20} | {'Genres':<30}")
print("-" * 120)

# Quét dữ liệu và áp dụng tất cả điều kiện lọc
for key, data in table.scan():
    try:
        # Lấy dữ liệu và giải mã
        title = data.get(b'CoBan:title', b'N/A').decode('utf-8', errors='ignore')
        type_ = data.get(b'CoBan:type', b'').decode('utf-8', errors='ignore')
        year = data.get(b'CoBan:release_year', b'').decode('utf-8', errors='ignore')
        country = data.get(b'ChiTiet:country', b'').decode('utf-8', errors='ignore')
        listed_in = data.get(b'ChiTiet:listed_in', b'').decode('utf-8', errors='ignore')

        # Áp dụng 3 điều kiện
        if (
            year > "2019" and                 # So sánh chuỗi an toàn với năm dạng 'YYYY'
            "United States" in country and
            type_ == "Movie" and
            "Dramas" in listed_in
        ):
            print(f"{key.decode():<8} | {title[:30]:<30} | {year:<6} | {type_[:10]:<10} | {country[:20]:<20} | {listed_in[:30]:<30}")
    except:
        continue

conn.close()
```

```python
#Retreive More than one Row
kxn.open()
tEmpDept2 = kxn.table('EmpDept2')
rowsDict = dict(tEmpDept2.rows([b'742866', b'897889']))
for k,v in rowsDict.items():
    print(k,v)
```

```python
#Retreive More than one Row, but only one column family
kxn.open()
tEmpDept2 = kxn.table('EmpDept2')
rows = tEmpDept2.rows([b'742866', b'897889'],columns=[b'Emp'])
for key, data in rows:
    print(key, data)
```

```python
#Filter
kxn.open()
tEmpDept2 = kxn.table('EmpDept2')
filter = b"SingleColumnValueFilter('Emp', 'Salary', <, 'binary:25000') AND SingleColumnValueFilter('Dept', 'DeptID', >, 'binary:30')"
for key, data in tEmpDept2.scan(
                                filter=filter):
    print(key, data)
kxn.close()
```
